```python
import statistics
import functions_salary as f_salary

# with open("employes_data.json", "r") as f:
#     data = json.load(f)
data = f_salary.o_JSON("employes_data.json")
stats_company = []
salaire_max_company = [0,0]
salaire_min_company = [10000,0]

#company = data["TechCorp"] #Liste contenant un dictionnaire des employes de la filiale
for company_name in data:
    company= data[company_name]
    for person in company:
        name = person.get("name")
        job = person.get("job")


        salaire = f_salary.compute_salary(person["weekly_hours_worked"], person["contract_hours"], person["hourly_rate"])
        
        person['salaire'] = float(salaire)
    # on va trier par salaire croissant
    sorted_company = sorted(company, key =lambda x : x['salaire'],reverse = True)

    # Step 3 : Calcul des stats des employes company
    salaire_min = float(min([d['salaire'] for d in company ]))
    salaire_mean = round(statistics.mean([d['salaire'] for d in company ]) , 2)

    salaire_max = float(max([d['salaire'] for d in company ]))
    if salaire_max > salaire_max_company[0]:
        salaire_max_company[0] = salaire_max
        salaire_max_company[1] = person['name']

    if salaire_min < salaire_min_company[0]:
        salaire_min_company[0] = salaire_min
        salaire_min_company[1] = person['name']



    stats_filiale = []
    stats_filiale.append({"salaire_max" : salaire_max})
    stats_filiale.append({"salaire_min" : salaire_min})
    stats_filiale.append({"salaire_mean" : salaire_mean})

    # Calcul des statistiques globales
    stats_company.extend(stats_filiale)


####################################################################################
    print(f"{company_name}")
    print('*'*70)

    #on def l'alignement du print
    l_nom = 10
    l_job = 18

    for person in sorted_company:
        print(f"{person['name'].ljust(l_nom)} | {person['job'].ljust(l_job)} | salaire mensuel : {str(person['salaire'])}€")

        #f_salary.print_result_filiale(company_name,stats_filiale)
    print('='*70)
    print(f"Statistique des salaires pour la filiale {company_name} : ")
    print(f"Salaire moyen : {stats_filiale[2]['salaire_mean']}€")
    print(f"Salaire minimum : {stats_filiale[1]['salaire_min']}€")
    print(f"Salaire maximum : {stats_filiale[0]['salaire_max']}€")
    print('='*70)

print('_'*70)
print(f"Statistique des salaires pour la compagnie mère : ")
valeur_salaires_mean = [dico['salaire_mean'] for dico in stats_company if 'salaire_mean' in dico]
print(f"Salaire moyen   : {round(statistics.mean(valeur_salaires_mean),2)}€")
print(f"Salaire minimum : {salaire_min_company[0]}€")
print(f"Salaire maximum : {salaire_max_company[0]}€")
```
